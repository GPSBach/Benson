```python
"""
Set Options
"""

# import libraries
import pandas as pd
import numpy as np
import os
import glob
import matplotlib.pyplot as plt
import matplotlib
from datetime import datetime
import pickle as pkl

# configuration options
%matplotlib inline
matplotlib.style.use("seaborn-muted")
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 50)
```

```python
"""
import pickle file
"""
with open("./data/filtered_data.pkl", 'rb') as picklefile: 
    filtered = pkl.load(picklefile)
```

```python
filtered.head()
```

```python
filtered.sort_values("ENTRY_FLUX",ascending=False).STATION.unique()
```

```python
"""
plot distribution of entries
"""
fig, ax = plt.subplots()
filtered[filtered['STATION']=='GRD CNTRL-42 ST'].ENTRY_FLUX.hist(ax=ax, bins=100, bottom=0)
#station.boxplot()
#ax.set_xscale('log')
```

```python
#filtered_by_station = filtered.groupby(['STATION']).ENTRY_FLUX.sum().sort_values().resetindex()
filtered.groupby(['STATION']).ENTRY_FLUX.sum().sort_values().reset_index()
```

```python
filtered
```

```python
filtered_station = filtered.groupby(['STATION']).agg({'ENTRY_FLUX':['sum', 'median'], 'EXIT_FLUX' : ['sum', 'median']})
filtered_station.columns = filtered_station.columns.map('_'.join)
```

```python
filtered_station.sort_values(['ENTRY_FLUX_sum'], ascending = False)
```

```python
filtered_station_date = filtered.groupby(['STATION', 'DATE']).agg({'ENTRY_FLUX':['sum', 'median'], 'EXIT_FLUX' : ['sum', 'median']})
filtered_station_date.columns = filtered_station_date.columns.map('_'.join)
```

```python
filtered_station_date.sort_values(['ENTRY_FLUX_sum'], ascending = False)
```
